```python
import pandas as pd
data=pd.read_csv("易联众样式数据.csv")
#print(base_data)
to_data=data.loc[:,['疾病代码（ICD10）']].drop_duplicates().reset_index()
trage_list=[]
for i in range(len(to_data['疾病代码（ICD10）'])):
    trage_list.append(i)
to_data['target']=trage_list
to_data=to_data.loc[:,['疾病代码（ICD10）','target']]
to_data.values
```

```python
base_data=data.loc[:,['患者性别','年龄','疾病代码（ICD10）']]
df=pd.merge(base_data,to_data,how='left',on='疾病代码（ICD10）')
df=df.loc[:,['患者性别','年龄','target']]
df.values
```
